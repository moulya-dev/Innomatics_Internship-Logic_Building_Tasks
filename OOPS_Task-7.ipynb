Problem 1: Smart Light Controller

```python
# Class to control smart light
class SmartLight:

    # Constructor to initialize light name and status
    def __init__(self, name):
        self.name = name
        self.status = "OFF"   # default status

    # Method to turn ON light
    def turn_on(self):
        self.status = "ON"

    # Method to turn OFF light
    def turn_off(self):
        self.status = "OFF"

    # Method to display status
    def display_status(self):
        print(self.name, "is", self.status)


# User input
light_name = input("Enter light name: ")
action = input("Enter action (ON/OFF): ")

# Create object
light = SmartLight(light_name)

# Perform action
if action == "ON":
    light.turn_on()
else:
    light.turn_off()

# Display result
light.display_status()
```

Problem 2: Employee ID Card System

```python
# Class to store employee details
class Employee:

    # Constructor
    def __init__(self, name, emp_id, department):
        self.name = name
        self.emp_id = emp_id
        self.department = department

    # Method to display ID card
    def display_id_card(self):
        print("\nEmployee ID Card")
        print("Name:", self.name)
        print("ID:", self.emp_id)
        print("Department:", self.department)


# User input
name = input("Enter employee name: ")
emp_id = input("Enter employee ID: ")
dept = input("Enter department: ")

# Create object
emp = Employee(name, emp_id, dept)

# Display ID card
emp.display_id_card()
```

Problem 3: Mobile Contact Record

```python
# Class to store contact details
class Contact:

    # Constructor
    def __init__(self, name, phone):
        self.name = name
        self.phone = phone

    # Method to display contact
    def display_contact(self):
        print("\nContact Saved")
        print("Name:", self.name)
        print("Phone:", self.phone)


# User input
name = input("Enter contact name: ")
phone = input("Enter phone number: ")

# Create object
contact = Contact(name, phone)

# Display contact
contact.display_contact()
```

Problem 4: Product Price Tag Generator

```python
# Class to store product details
class Product:

    # Constructor
    def __init__(self, name, price):
        self.name = name
        self.price = price

    # Method to display price tag
    def display_price_tag(self):
        print("\nProduct:", self.name)
        print("Price: ₹" + str(self.price))


# User input
name = input("Enter product name: ")
price = input("Enter price: ")

# Create object
product = Product(name, price)

# Display price tag
product.display_price_tag()
```

Problem 5: Movie Rating Display System

```python
# Class to store movie details
class Movie:

    # Constructor
    def __init__(self, name, rating):
        self.name = name
        self.rating = rating

    # Method to display movie details
    def display_movie(self):
        print("\nMovie:", self.name)
        print("Rating:", self.rating, "/ 5")


# User input
name = input("Enter movie name: ")
rating = input("Enter rating: ")

# Create object
movie = Movie(name, rating)

# Display movie details
movie.display_movie()
```

Problem 6: Delivery Address Manager

```python
# Class to store delivery details
class Delivery:

    # Constructor
    def __init__(self, customer, address):
        self.customer = customer
        self.address = address

    # Method to display delivery details
    def display_delivery(self):
        print("\nDelivery Details")
        print("Customer:", self.customer)
        print("Address:", self.address)


# User input
customer = input("Enter customer name: ")
address = input("Enter address: ")

# Create object
delivery = Delivery(customer, address)

# Display delivery details
delivery.display_delivery()
```
