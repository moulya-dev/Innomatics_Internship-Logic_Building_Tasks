Unique Words in a Sentence

```python
# Taking input from user
sentence = input("Enter sentence: ")

# Split sentence into words using space
words = sentence.split()

# Convert list to set to remove duplicates
unique_words = set(words)

# Display results
print("Unique words count:", len(unique_words))
print("Unique words:", unique_words)
```

 Highest Salary from Employee Data

```python
# Employee dictionary
employees = {
    "Ravi": 75000,
    "Anita": 68000,
    "Kiran": 72000
}

# Initialize highest salary variables
highest_employee = None
highest_salary = 0

# Loop through dictionary to find highest salary
for name, salary in employees.items():
    if salary > highest_salary:
        highest_salary = salary
        highest_employee = name

# Display result
print("Highest Salary:", highest_employee, "-", highest_salary)
```

Find Maximum and Minimum Values

```python
numbers = [45, 22, 89, 10, 66]

# Initialize max and min with first element
maximum = numbers[0]
minimum = numbers[0]

# Loop through numbers
for num in numbers:
    if num > maximum:
        maximum = num
    if num < minimum:
        minimum = num

# Display results
print("List:", numbers)
print("Max:", maximum)
print("Min:", minimum)
```

Count Products Above a Price Threshold

```python
prices = [450, 1200, 899, 1500, 300]

count = 0  # Counter for prices above 1000

# Loop through prices
for price in prices:
    if price > 1000:
        count += 1

# Display result
print("Products above 1000:", count)
```

 Calculate Attendance Percentage

```python
attendance = ["P", "P", "A", "P", "P"]

# Count number of Present days
present_count = attendance.count("P")

# Total number of days
total_days = len(attendance)

# Calculate percentage
attendance_percentage = (present_count / total_days) * 100

# Display result
print("Attendance Percentage:", attendance_percentage)
```

 Count Character Frequency

```python
# Take string input from user
string = input("Enter a string: ")

# Create empty dictionary to store frequency
char_frequency = {}

# Loop through each character in string
for char in string:
    # If character already exists, increment count
    if char in char_frequency:
        char_frequency[char] += 1
    else:
        # If character appears first time, initialize with 1
        char_frequency[char] = 1

# Display result
print("Character Frequency:", char_frequency)
```

Convert List to Tuple

```python
# Take user input (space separated numbers)
user_input = input("Enter numbers separated by space: ")

# Convert string input into list of integers
numbers_list = list(map(int, user_input.split()))

# Convert list to tuple
numbers_tuple = tuple(numbers_list)

# Display result
print("Tuple:", numbers_tuple)
```

Check if a Key Exists in a Dictionary

```python
# Create empty dictionary
employees = {}

# Number of employees to enter
n = int(input("Enter number of employees: "))

# Take employee details from user
for i in range(n):
    name = input("Enter employee name: ")
    salary = int(input("Enter employee salary: "))
    employees[name] = salary

# Key to check
key_to_check = input("Enter employee name to check: ")

# Check if key exists
if key_to_check in employees:
    print("Employee exists")
else:
    print("Employee does not exist")
```

Calculate Average Marks

```python
# Take number of subjects
n = int(input("Enter number of subjects: "))

marks = []

# Take marks input from user
for i in range(n):
    mark = float(input(f"Enter marks for subject {i+1}: "))
    marks.append(mark)

# Calculate total marks
total = sum(marks)

# Calculate average
average = total / len(marks)

# Display result
print("Average Marks:", average)
```
