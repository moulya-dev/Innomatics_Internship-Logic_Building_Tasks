Problem 1: Mobile Recharge Validation System

```python
def validate_recharge():
    valid_plans = [199, 299, 399, 599]
    
    while True:
        amount = int(input("Enter recharge amount: ₹"))
        
        # Check minimum recharge rule
        if amount < 50:
            print("Recharge amount must be at least ₹50. Try again.")
            continue
        
        # Check valid plans
        if amount in valid_plans:
            print("Recharge successful!")
            break
        else:
            print("Invalid plan. Available plans:", valid_plans)
            print("Please try again.\n")

# Call function
validate_recharge()
```

Problem 2: Inventory Reorder Alert System

```python
def check_inventory(products):
    for product, stock in products.items():
        if stock < 15:
            print(f"{product}: Reorder Alert  (Stock = {stock})")
        else:
            print(f"{product}: Stock OK  (Stock = {stock})")

# Sample dictionary
inventory = {
    "Laptop": 20,
    "Mouse": 10,
    "Keyboard": 5,
    "Monitor": 18
}

check_inventory(inventory)
```

Problem 3: Student Result Processing System

```python
def process_result(marks):
    total = 0
    
    # Calculate total using loop
    for mark in marks:
        total += mark
    
    average = total / len(marks)
    
    print("Average Marks:", average)
    
    if average >= 50:
        return "Pass"
    else:
        return "Fail"

# Example usage
marks_list = list(map(int, input("Enter numbers separated by space: ").split()))
result = process_result(marks_list)
print("Result:", result)
```

Problem 4: Cab Fare Estimator with Retry Option

```python
def calculate_fare(distance, peak):
    base_fare = 50
    per_km_rate = 12
    
    fare = base_fare + (distance * per_km_rate)
    
    # Add peak charge
    if peak.lower() == "yes":
        fare += fare * 0.25  # 25% extra
    
    return fare


while True:
    distance = float(input("Enter distance (in km): "))
    peak = input("Is it peak hour? (yes/no): ")
    
    total_fare = calculate_fare(distance, peak)
    print("Total Fare: ₹", total_fare)
    
    retry = input("Do you want to calculate again? (yes/no): ")
    if retry.lower() != "yes":
        break
```

Problem 5: Employee Attendance Eligibility Checker

```python
def check_eligibility(attendance_list):
    total_days = len(attendance_list)
    present_days = 0
    
    # Count present days
    for day in attendance_list:
        if day == "P":
            present_days += 1
    
    percentage = (present_days / total_days) * 100
    print("Attendance Percentage:", percentage)
    
    if percentage >= 75:
        return "Eligible"
    else:
        return "Not Eligible"

# Example
attendance = ["P", "P", "A", "P", "A", "P", "P", "P"]
status = check_eligibility(attendance)
print("Status:", status)
```

Problem 6: Password Strength Checker

```python
def check_password(password):
    has_digit = False
    has_special = False
    special_chars = "@#$"
    
    # Check length
    if len(password) < 8:
        return "Weak Password (Minimum 8 characters required)"
    
    # Loop through characters
    for char in password:
        if char.isdigit():
            has_digit = True
        if char in special_chars:
            has_special = True
    
    # Validate conditions
    if has_digit and has_special:
        return "Strong Password"
    else:
        return "Weak Password (Must include digit and special character @#$)"

# Example
user_password = input("Enter password: ")
print(check_password(user_password))
```
