Problem 1: Smart Parking Lot Management System

```python
# Function to manage smart parking lot
def smart_parking_system(capacity, logs):

    parked = 0          # current parked vehicles
    peak = 0            # peak parking usage

    # Loop through vehicle logs
    for log in logs:

        if log == "IN":
            parked += 1

        elif log == "OUT":
            if parked > 0:
                parked -= 1

        # Update peak usage
        if parked > peak:
            peak = parked

    # Display results
    print("Currently Parked Vehicles:", parked)
    print("Peak Parking Usage:", peak)

    # Check capacity alert
    if parked >= capacity:
        print("Parking Status: FULL / ALERT")
    else:
        print("Parking Status: Available")


# User input
capacity = int(input("Enter parking capacity: "))

logs_input = input("Enter vehicle logs separated by space (IN/OUT): ")
logs = logs_input.split()

# Call function
smart_parking_system(capacity, logs)
```

Problem 2: Online Food Delivery Time Estimator

```python
# Function to estimate delivery time
def estimate_delivery_time(distance, traffic, weather):

    base_time = distance * 5   # assume 5 minutes per km

    # Traffic delay
    if traffic.lower() == "low":
        traffic_delay = 0
    elif traffic.lower() == "medium":
        traffic_delay = 10
    else:  # High
        traffic_delay = 20

    # Weather delay
    if weather.lower() == "clear":
        weather_delay = 0
    elif weather.lower() == "rainy":
        weather_delay = 15
    else:  # storm/fog
        weather_delay = 25

    total_time = base_time + traffic_delay + weather_delay

    print("Estimated Delivery Time:", total_time, "minutes")


# User input
distance = float(input("Enter distance (km): "))
traffic = input("Enter traffic level (Low/Medium/High): ")
weather = input("Enter weather condition (Clear/Rainy/Storm): ")

estimate_delivery_time(distance, traffic, weather)
```

Problem 3: Movie Theatre Seat Occupancy Analyzer

```python
# Function to analyze seat occupancy
def analyze_occupancy(total_seats, booked_list):

    booked = len(booked_list)

    occupancy = (booked / total_seats) * 100

    print("Occupancy:", round(occupancy), "%")

    # Determine status
    if occupancy == 100:
        print("Show Status: Housefull")
        print("Suggestion: Open additional shows")

    elif occupancy >= 75:
        print("Show Status: Almost Full")

    else:
        print("Show Status: Available")


# User input
total_seats = int(input("Enter total seats: "))

booked_count = int(input("Enter number of booked seats: "))

# create list of booked seats
booked_list = [1] * booked_count

analyze_occupancy(total_seats, booked_list)
```

Problem 4: Cloud Server Load Classification System

```python
# Function to classify server load
def classify_server_load(cpu_readings):

    total = 0

    # Calculate sum
    for reading in cpu_readings:
        total += reading

    average = total / len(cpu_readings)

    print("Average CPU Load:", round(average), "%")

    # Classification
    if average < 50:
        print("Server Status: Normal")

    elif average <= 80:
        print("Server Status: Warning")

    else:
        print("Server Status: Critical")


# User input
cpu_input = input("Enter CPU readings separated by space: ")

cpu_readings = list(map(int, cpu_input.split()))

classify_server_load(cpu_readings)
```

Problem 5: Smart Classroom Resource Usage Monitor

```python
# Function to monitor resource usage
def monitor_resources(resources):

    overused = []

    # Assume overuse if usage > 8 hours
    for resource, hours in resources.items():

        if hours > 8:
            overused.append(resource)

    # Output
    if overused:
        print("Overused Resources:", ", ".join(overused))
        print("Energy Alert: Yes")
    else:
        print("Overused Resources: None")
        print("Energy Alert: No")


# User input
resources = {}

n = int(input("Enter number of resources: "))

for i in range(n):

    name = input("Enter resource name: ")
    hours = int(input("Enter usage hours: "))

    resources[name] = hours

monitor_resources(resources)
```

Problem 6: Online Event Registration Capacity Controller

```python
# Function to manage registrations
def manage_registrations(capacity, registrations):

    if registrations <= capacity:

        print("Confirmed Registrations:", registrations)
        print("Waitlisted Users: 0")
        print("Registration Status: Open")

    else:

        waitlist = registrations - capacity

        print("Confirmed Registrations:", capacity)
        print("Waitlisted Users:", waitlist)
        print("Registration Status: Closed")


# User input
capacity = int(input("Enter event capacity: "))
registrations = int(input("Enter number of registrations: "))

manage_registrations(capacity, registrations)
```
