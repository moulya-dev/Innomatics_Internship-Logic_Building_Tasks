Problem Statement 1: Employee Performance Bonus Eligibility

```python
# Dictionary containing employee names and their performance scores
employees = {
    "Ravi": 92,
    "Anita": 88,
    "Kiran": 92,
    "Suresh": 85
}

# Step 1: Identify the highest performance score
highest_score = max(employees.values())

# Step 2: Handle ties by selecting all employees with the highest score
top_performers = [name for name, score in employees.items() if score == highest_score]

# Step 3: Display the result
print(f"Top Performers Eligible for Bonus: {', '.join(top_performers)} (Score: {highest_score})")
```

Problem Statement 2: Search Query Keyword Analysis

```python
import string

# Input search query
query = "Buy mobile phone buy phone online"

# Step 1: Convert to lowercase
query = query.lower()

# Step 2: Remove punctuation
query = query.translate(str.maketrans('', '', string.punctuation))

# Step 3: Split into words
words = query.split()

# Step 4: Count frequency of each word
frequency = {}

for word in words:
    frequency[word] = frequency.get(word, 0) + 1

# Step 5: Display only words searched more than once
repeated_keywords = {word: count for word, count in frequency.items() if count > 1}

print(repeated_keywords)
```

Problem Statement 3: Sensor Data Validation

```python
sensor_readings = [3, 4, 7, 8, 10, 12, 5]

valid_readings = []

# Step 1: Loop through readings with index
for hour, value in enumerate(sensor_readings):
    
    # Step 2: Check if reading is even (valid)
    if value % 2 == 0:
        valid_readings.append((hour, value))

# Step 3: Display result
print("Valid Sensor Readings (Hour, Value):")
print(valid_readings)
```

Problem Statement 4: Email Domain Usage Analysis

```python
emails = [
    "ravi@gmail.com",
    "anita@yahoo.com",
    "kiran@gmail.com",
    "suresh@gmail.com",
    "meena@yahoo.com"
]

domain_count = {}

# Step 1: Extract domain and count occurrences
for email in emails:
    domain = email.split('@')[1]
    domain_count[domain] = domain_count.get(domain, 0) + 1

# Step 2: Calculate percentage usage
total_emails = len(emails)

print("Email Domain Usage Percentage:")
for domain, count in domain_count.items():
    percentage = (count / total_emails) * 100
    print(f"{domain}: {percentage:.0f}%")
```

Problem Statement 5: Sales Spike Detection

```python
sales = [1200, 1500, 900, 2200, 1400, 3000]

# Step 1: Calculate average daily sales
average_sales = sum(sales) / len(sales)

# Step 2: Define spike threshold (30% above average)
threshold = average_sales * 1.30

print("Sales Spike Days:")

# Step 3: Detect spike days
for day, value in enumerate(sales, start=1):
    if value > threshold:
        print(f"Day {day}: {value}")
```

Problem Statement 6: Duplicate User ID Detection

```python
user_ids = ["user1", "user2", "user1", "user3", "user1", "user3"]

id_count = {}

# Step 1: Count occurrences of each user ID
for user in user_ids:
    id_count[user] = id_count.get(user, 0) + 1

# Step 2: Display only duplicates
for user, count in id_count.items():
    if count > 1:
        print(f"{user} → {count} times")
```
