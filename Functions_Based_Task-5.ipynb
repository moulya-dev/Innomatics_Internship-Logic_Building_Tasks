Problem 1: Social Media – Post Engagement Analyzer

```python
# Function to analyze post engagement
def analyze_post_engagement():
    
    # Taking number of posts as input
    n = int(input("Enter number of posts: "))
    
    likes_list = []
    
    # Loop to collect likes for each post
    for i in range(n):
        likes = int(input(f"Enter likes for post {i+1}: "))
        likes_list.append(likes)
    
    # Calculating total likes
    total_likes = 0
    for likes in likes_list:
        total_likes += likes
        
    # Checking viral condition
    if total_likes >= 1000:
        status = "Viral Post"
    else:
        status = "Normal Engagement"
    
    print("\nTotal Likes:", total_likes)
    print("Post Status:", status)

# Function call
analyze_post_engagement()
```

Problem 3: Agriculture – Rainfall Adequacy Checker

```python
# Function to check rainfall adequacy
def rainfall_checker():
    
    n = int(input("Enter number of rainfall records: "))
    rainfall_data = []
    
    # Collect rainfall data
    for i in range(n):
        rain = float(input(f"Enter rainfall for day {i+1}: "))
        rainfall_data.append(rain)
    
    required_level = float(input("Enter required rainfall level: "))
    
    # Calculate total rainfall
    total = 0
    for rain in rainfall_data:
        total += rain
    
    # Calculate average rainfall
    average = total / len(rainfall_data)
    
    # Condition check
    if average >= required_level:
        status = "Adequate Rainfall"
    else:
        status = "Inadequate Rainfall"
    
    print("\nAverage Rainfall:", average)
    print("Rainfall Status:", status)

# Function call
rainfall_checker()
```

Problem 4: Social Media – Duplicate Account Detection

```python
# Function to detect duplicate usernames
def detect_duplicates():
    
    n = int(input("Enter number of usernames: "))
    usernames = []
    
    # Collect usernames
    for i in range(n):
        name = input(f"Enter username {i+1}: ")
        usernames.append(name)
    
    # Using set to detect duplicates
    if len(usernames) != len(set(usernames)):
        result = "Yes"
    else:
        result = "No"
    
    print("\nDuplicate Accounts Found:", result)

# Function call
detect_duplicates()
```

Problem 6: Agriculture – Premium Crop Price Filter

```python
# Function to filter premium crops
def premium_crop_filter():
    
    n = int(input("Enter number of crop prices: "))
    prices = []
    
    # Collect crop prices
    for i in range(n):
        price = int(input(f"Enter price {i+1}: "))
        prices.append(price)
    
    premium = []
    
    # Filtering prices above ₹2000
    for price in prices:
        if price > 2000:
            premium.append(price)
    
    print("\nPremium Crops:", premium)

# Function call
premium_crop_filter()
```

Problem 9: E-Learning – Student Attendance Eligibility System

```python
# Function to check attendance eligibility
def attendance_checker():
    
    n = int(input("Enter number of classes conducted: "))
    attendance_list = []
    
    # 1 = Present, 0 = Absent
    for i in range(n):
        status = int(input(f"Enter attendance for class {i+1} (1=Present, 0=Absent): "))
        attendance_list.append(status)
    
    total = 0
    
    # Calculating total present days
    for day in attendance_list:
        total += day
    
    # Calculate attendance percentage
    percentage = (total / len(attendance_list)) * 100
    
    # Eligibility condition
    if percentage >= 75:
        status = "Eligible"
    else:
        status = "Not Eligible"
    
    print("\nAttendance Percentage:", percentage)
    print("Exam Eligibility:", status)

# Function call
attendance_checker()
```

Problem 10: Smart Electricity Bill Analyzer

```python
# Function to calculate electricity bill
def electricity_bill():
    
    units = int(input("Enter units consumed: "))
    bill = 0
    
    # Billing slabs
    if units <= 100:
        bill = units * 3
    elif units <= 200:
        bill = (100 * 3) + (units - 100) * 5
    else:
        bill = (100 * 3) + (100 * 5) + (units - 200) * 7
    
    # Usage classification
    if bill < 500:
        usage = "Low Usage"
    elif bill <= 1500:
        usage = "Moderate Usage"
    else:
        usage = "High Usage"
    
    print("\nUnits Consumed:", units)
    print("Total Bill:", bill)
    print("Usage Level:", usage)

# Function call
electricity_bill()
```
